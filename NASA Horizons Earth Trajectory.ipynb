```python
import numpy as np
import matplotlib.pyplot as plt

filename = "EARTH_horizons_results.txt"

# Initialize an empty list to store the positions
positions = []

# Open the file and read the data
with open(filename, 'r') as file:
    # Flag to indicate whether we are in the data section
    read_data = False
    for line in file:
        # If we encounter $$SOE, start reading data
        if "$$SOE" in line:
            read_data = True
            continue
        # If we encounter $$EOE, stop reading data
        elif "$$EOE" in line:
            read_data = False
        # If we are in the data section, read the X, Y, Z coordinates
        elif read_data:
            parts = line.split(',')
            x, y, z = float(parts[2]), float(parts[3]), float(parts[4])
            positions.append([x, y, z])

# Convert the list to a numpy array
positions = np.array(positions, dtype=np.float64)

# Plot X against Y
plt.figure(figsize=(10, 10))
plt.plot(positions[:, 0], positions[:, 1], label="NASA")
plt.xlabel("x (AU)")
plt.ylabel("y (AU)")
plt.title("Earth's Trajectory")
plt.legend()
plt.show()
```
